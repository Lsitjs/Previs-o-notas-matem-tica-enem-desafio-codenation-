Gerando as features que mais correlacionam

```python
features=['NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_MT','NU_NOTA_REDACAO','NU_NOTA_COMP1','NU_NOTA_COMP2','NU_NOTA_COMP3','NU_NOTA_COMP4','NU_NOTA_COMP5']
#trocando os valores nan por 0
featuresnonull=enem_train[features].replace(np.NAN, 0)

x=featuresnonull.drop(columns=['NU_NOTA_MT'])

y=featuresnonull['NU_NOTA_MT'].replace(np.NAN, 0)

x_features=['NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_REDACAO','NU_NOTA_COMP1','NU_NOTA_COMP2','NU_NOTA_COMP3','NU_NOTA_COMP4','NU_NOTA_COMP5']

x_test=enem_test[x_features].replace(np.NAN, 0)
```

Fazendo a regressão

```python
lr= LinearRegression()

model=lr.fit(x,y)
```

Fazendo a predição

```python
resultado=model.predict(x_test)
```

```python
resultado
```

Gerando o Dataframe o "answer.csv" e trocando os valores negativos do resultado por 0

```python
resposta=pd.DataFrame()

resposta['NU_INSCRICAO']=enem_test['NU_INSCRICAO']

resposta['NU_NOTA_MT'] = np.around(resultado,2)

resposta['NU_NOTA_MT'][resposta['NU_NOTA_MT'] < 0] = 0

resposta.to_csv('answer.csv', index=False, header=True)
```
